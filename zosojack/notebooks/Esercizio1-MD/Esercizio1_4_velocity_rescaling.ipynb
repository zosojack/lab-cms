## Svolgimento PUNTO 4

### 0) Importazioni e dati

```python
import numpy as np
from CMS.MolecularDynamics.CrystalStructure import CrystalStructure as Crystal
from CMS.MolecularDynamics.CrystalDynamics import CrystalDynamics
```

```python
# Nome del file (numero di atomi da studiare: 256)
filename = '../../../data/pos-fcc100-minime/C_s=5e-4~F_t=5e-4~max_s=50k/fcc100a256_1.txt'
```

#### Dati globali

```python
# boundaries della cella da studiare (crystal slab)
# pbc = [16.6416, 16.6416, np.inf]  # in √Ö
pbc = [16.6416, 16.6416, 166_416]  # in √Ö (z enorme (*10_000))

# temperature da studiare
temperatura = [300, 850] # in K, cerco una temperatura finale ~850 K (rescale velocit√†)

# massa atomica dell'atomo considerato (argento)
silver_mass = 108

# time step ottimale per 1000K
time_step = {
    300: 6e-15,
    850: 4e-15 
}

# il numero di step √® dato dal tempo totale (300ps) diviso il time step
tempo_totale = 300e-12
n_steps = {T: int(tempo_totale / time_step[T]) for T in temperatura}

# tempo di termalizzazione (ogni valore precedente va buttato)
t_th = 3e-12
```

#### Inizializzazione del cristallo

```python
# Inizializzo il cristallo dal file minimizzato con steepest-descent
cristallo = Crystal.from_file(filename)
# Setto il raggio di cutoff E la giunzione polinomiale
cristallo.set_R_C(4.5)
cristallo.set_R_P(4.2)
cristallo.set_R_V(5.0)
# Setto le condizioni al contorno periodiche (vanno settate DOPO R_C)
cristallo.set_pbc(pbc)
```

#### Aggiunta di un atomo in superficie al centro e a +2.4 √Ö dal massimo z

```python
# prendo il massimo z di tutti gli atomi
massimo_z = np.max(cristallo.positions[:, 2])
# calcolo il centro del cristallo nelle direzioni x e y
centro_x, centro_y, _ = cristallo.crystal_center
# posiziono l'atomo ad una distanza di 2.4 √Ö sopra il massimo z
cristallo.add_atom([centro_x, centro_y, massimo_z + 2.4])

# 16.6416 √® dimensione della cella, poi traslo in base all'origine
origine = np.min(cristallo.positions, axis=0)
#cristallo.add_atom([16.6416/2 + origine[0], 16.6416/2 + origine[1], massimo_z + 2.4])
```

### 1) Simulazione di 300ps a T = 850 K

```python
from CMS.MolecularDynamics.io import AtomTracker, XYZwriter
# Inizializzo il tracker per l'atomo aggiunto (ultimo atomo)
adatom_tracker_unbounded = {
    850: AtomTracker(index=-1, output_file='output/1_4/Esercizio1_4_850_vr/adatom_track_unbounded.txt'),
    300: AtomTracker(index=-1, output_file='output/1_4/Esercizio1_4_300_vr/adatom_track_unbounded.txt'),
}
adatom_tracker_periodic = {
    850: AtomTracker(index=-1, output_file='output/1_4/Esercizio1_4_850_vr/adatom_track_periodic.txt', pbc_option = 'periodic'),
    300: AtomTracker(index=-1, output_file='output/1_4/Esercizio1_4_300_vr/adatom_track_periodic.txt', pbc_option = 'periodic'),
}
# Inizializzo lo scrittore XYZ per salvare la dinamica
xyz_writers = {
    850: XYZwriter(output_folder='output/1_4/Esercizio1_4_850_vr/', dt=time_step[850], dump_interval=200, pbc_option='periodic'), 
    300: XYZwriter(output_folder='output/1_4/Esercizio1_4_300_vr/', dt=time_step[300], dump_interval=200)
}   
```

Per 256 atomi, due temperature, con scrittura dei file: 60 secondi

```python
result = {}
```

```python
for T in temperatura:
    print(f"üöÄ Avvio simulazione a T={T} K")
    # Raccolgo le energie medie e le dev std
    simulazione = CrystalDynamics(cristallo.copy(), 
                        atomic_mass=silver_mass, 
                        dt=time_step[T], 
                        temp_ini=T, 
                        atom_tracker=[adatom_tracker_unbounded[T], adatom_tracker_periodic[T]], 
                        xyz_writer=xyz_writers[T])  
    simulazione.set_seed(12316543)

    # stavolta output True per visualizzare con ovito
    result[T] = simulazione.run_dynamics(n_steps=n_steps[T], 
                                         t_th=t_th,
                                         rescale_velocity=True,
                                         debug=False)
```

### 2) Analisi e plot

Alcuni dati della simulazione

```python
for T in temperatura:
    print(f"Risultati per T={T} K")
    result[T].summary()
```

#### 2.1) Temperatura finale

```python
print(f"{result[850].mean_temp} ¬± {result[850].std_temp} K")
```

Provo a plottare anche un fotogramma della superficie. Non √® rappresentativo, poich√© non se ne vede l'evoluzione nel tempo, ma pu√≤ essere carino.

L'adatom viene posto al centro, cio√® in una posizione instabile. Quindi molto velocemente, subito nei primi step, va a spostarsi in un 'buco', infilato tra quattro atomi. Succede talmente presto, che la .gif di Ovito parte da una configurazione in cui si √® gi√† spostato.

```python
import matplotlib.pyplot as plt

data1 = np.loadtxt("output/1_4/Esercizio1_4_850_vr/adatom_track_unbounded.txt", skiprows=2, delimiter="\t")
x1 = data1[:, 1]
y1 = data1[:, 2]

data2 = np.loadtxt("output/1_4/Esercizio1_4_300_vr/adatom_track_unbounded.txt", skiprows=2, delimiter="\t")
x2 = data2[:, 1]
y2 = data2[:, 2]
```

```python
# metto due grafici per 850K e 300K uno accanto all'altro 
fig, ax = plt.subplots(1, 2, figsize=(12, 5), dpi=200)

# Creiamo un array di indici per il colore
time_indices1 = np.arange(len(x1)) * (time_step[850] * 1e12)  # Convertiamo gli step temporali in picosecondi

# Usiamo scatter
ax[0].scatter(x1, y1, c=time_indices1, cmap='plasma', s=0.5, linewidth=0, alpha=0.5)

ax[0].set_xlabel('x (√Ö)')
ax[0].set_ylabel('y (√Ö)')
ax[0].set_title("(100) | unbounded | T=850K", fontweight='bold')
ax[0].axis('equal') # Per non distorcere la geometria

for sup in [atomi_superficiali]: 
    ax[0].scatter(sup[:, 0], sup[:, 1], s=10, alpha=0.5)

cbar1 = plt.colorbar(ax[0].collections[0], ax=ax[0])
cbar1.set_label('Tempo (ps)')

# Creiamo un array di indici per il colore
time_indices2 = np.arange(len(x2)) * (time_step[300] * 1e12)  # Convertiamo gli step temporali in picosecondi

# Usiamo scatter. 
# c=time_indices dice di usare l'indice per il colore
# s=0.5 rende i punti piccolissimi
# linewidth=0 toglie il bordo ai punti per evitare che si sovrappongano troppo
ax[1].scatter(x2, y2, c=time_indices2, cmap='plasma', s=0.5, linewidth=0, alpha=0.5)

ax[1].set_xlabel('x (√Ö)')
ax[1].set_ylabel('y (√Ö)')
ax[1].set_title("(100) | unbounded | T=300K", fontweight='bold')
ax[1].axis('equal') # Per non distorcere la geometria

for sup in [atomi_superficiali]: 
    ax[1].scatter(sup[:, 0], sup[:, 1], s=10, alpha=0.5)

cbar2 = plt.colorbar(ax[1].collections[0], ax=ax[1])
cbar2.set_label('Tempo (ps)')

plt.tight_layout()
plt.show()
```
